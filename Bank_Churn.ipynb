# Task 2

```python
# identify the null values and perform imputation
```

```python
movie.isnull().sum()/len(movie.index)*100 
# columns homepage has more than 64% of null values, dropping the column
```

```python
movie.drop(columns=['homepage'], axis=1, inplace=True)
```

```python
#identify and display all the production companies with their count
```

```python
import json
new_movie = movie.copy()
#storing the json format in list
dict_columns = ['genres', 'keywords', 'production_companies','production_countries','spoken_languages']
```

```python
#sort the data in descending order based on revenue and filter top 100 movies, find the measures of central tendency for 
#budget
#revenue
#runtime
#also perform outlier analysis for 3 columns
```

```python
top_100_movies = movie.sort_values(by='revenue', ascending=False)[:100]
top_100_mct = top_100_movies[['budget','revenue','runtime']]
```

```python
print(top_100_mct.describe())
print('----'*20)
print('Mode')
print(top_100_mct.mode().iloc[0])
```

```python
#outlliers
plt.figure(figsize=(10,6))
sns.boxplot(data=top_100_mct)
```

```python
#from the graph we can see there are few ouliers in revenue, and 1 outlier in budget and runtime doesn't have any outlier
```
